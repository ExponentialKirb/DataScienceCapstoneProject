```python
# Task 2: Mark the success/failed launches for each site on the map
# Initialize a folium map centered around a location
site_map = folium.Map(location=[28.5721, -80.648], zoom_start=7)

# Initialize empty dictionaries to store success and total launch counts for each site
success_counts = {}
total_counts = {}

# Iterate through spacex_df DataFrame
for index, row in spacex_df.iterrows():
    # Extract launch site coordinates and outcome
    site_coords = (row['Lat'], row['Long'])  # Convert list to tuple
    outcome = row['class']
    
    # Check if the launch site is already in the dictionaries
    if site_coords in success_counts:
        # Increment success count if outcome is 1 (successful)
        if outcome == 1:
            success_counts[site_coords] += 1
        # Increment total count
        total_counts[site_coords] += 1
    else:
        # Initialize success count and total count
        if outcome == 1:
            success_counts[site_coords] = 1
        else:
            success_counts[site_coords] = 0
        total_counts[site_coords] = 1

# Iterate through launch_sites_df DataFrame
for index, row in launch_sites_df.iterrows():
    # Extract launch site coordinates and name
    site_coords = (row['Lat'], row['Long'])  # Convert list to tuple
    site_name = row['Launch Site']
    
    # Calculate success rate for the launch site
    if site_coords in success_counts:
        success_rate = success_counts[site_coords] / total_counts[site_coords]
    else:
        success_rate = 0.0
    
    # Create CircleMarker object with the launch site coordinates, popup label, and color based on success rate
    circle_marker = folium.CircleMarker(location=site_coords, radius=10, 
                                        popup=folium.Popup(f"{site_name}<br>Success Rate: {success_rate:.2f}", parse_html=True),
                                        color='green' if success_rate > 0.5 else 'red', fill=True, fill_opacity=0.6)
    
    # Add CircleMarker object to the map
    site_map.add_child(circle_marker)

# Display the map
site_map
```

```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```

```python
# Function to calculate distance between two points using Haversine formula
def calculate_distance(lat1, lon1, lat2, lon2):
    # Convert latitude and longitude from degrees to radians
    lat1, lon1, lat2, lon2 = map(radians, [lat1, lon1, lat2, lon2])

    # Haversine formula
    dlon = lon2 - lon1
    dlat = lat2 - lat1
    a = sin(dlat/2)**2 + cos(lat1) * cos(lat2) * sin(dlon/2)**2
    c = 2 * atan2(sqrt(a), sqrt(1-a))
    distance = 6371 * c  # Radius of Earth in km
    return distance

# Coordinates of the closest coastline
coastline_lat = 28.56367
coastline_lon = -80.57163

# Initialize variables to store nearest launch site and its distance to the coastline
nearest_launch_site = None
nearest_distance = float('inf')  # Initialize with infinity

# Find the nearest launch site to the coastline
for index, row in launch_sites_df.iterrows():
    distance_to_coastline = calculate_distance(row['Lat'], row['Long'], coastline_lat, coastline_lon)
    if distance_to_coastline < nearest_distance:
        nearest_distance = distance_to_coastline
        nearest_launch_site = (row['Lat'], row['Long'])

# Add MousePosition control to the site_map
folium.plugins.MousePosition().add_to(site_map)

# Iterate through launch_sites_df DataFrame
for index, row in launch_sites_df.iterrows():
    # Calculate distance between launch site and coastline
    distance_to_coastline = calculate_distance(row['Lat'], row['Long'], coastline_lat, coastline_lon)
    
    # Create Marker object for launch site
    launch_site_marker = folium.Marker(location=[row['Lat'], row['Long']], 
                                       popup=f"Distance to coastline: {distance_to_coastline:.2f} km")
    
    # Add Marker to site_map
    site_map.add_child(launch_site_marker)
    
    # Draw PolyLine between nearest launch site and coastline
    polyline = folium.PolyLine(locations=[nearest_launch_site, [coastline_lat, coastline_lon]], color='blue')
    site_map.add_child(polyline)

# Mark down the closest coastline point on the map
folium.Marker(location=[coastline_lat, coastline_lon], icon=folium.Icon(color='blue')).add_to(site_map)

# Display the map
site_map
```
